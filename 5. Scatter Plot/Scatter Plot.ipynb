```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp= [14.2,16.4,11.9,15.2,18.5,22.1,19.4]
icecream_sales=[215,325,185,332,406,522,412]
tea_sales=[80,65,89,70,56,45,43]
```

## 1. Draw A Scatter Plot With Title, Xlabel, And YLabel

```python
plt.scatter(temp,icecream_sales)
plt.title('Temp Vs. Iceream Sales')
plt.xlabel('Temprature')
plt.ylabel('Ice Cream Sales')
plt.show()
```

## 2. Change Size of The-Dots

```python
plt.scatter(temp,icecream_sales,s=100)
plt.title('Temp Vs. Iceream Sales')
plt.xlabel('Temprature')
plt.ylabel('Ice Cream Sales')
plt.show()
```

## 3. Change Color of The-Dots

```python
plt.scatter(temp,icecream_sales,s=100,color="r")
plt.title('Temp Vs. Iceream Sales')
plt.xlabel('Temprature')
plt.ylabel('Ice Cream Sales')
plt.show()
```

## 4. Change The Marker

https://matplotlib.org/stable/api/markers_api.html

```python
plt.scatter(temp,icecream_sales,s=100,marker="*",color="r")
plt.title('Temp Vs. Iceream Sales')
plt.xlabel('Temprature')
plt.ylabel('Ice Cream Sales')
plt.show()
```

## 5. Change The Edge Color of The Marker Also change Line Width

```python
plt.scatter(temp,icecream_sales,s=100,marker="*",color="r",edgecolor='b',
           linewidths=1.5)
plt.title('Temp Vs. Iceream Sales')
plt.xlabel('Temprature')
plt.ylabel('Ice Cream Sales')
plt.show()
```

## 6. Change Alpha Parameter

```python
plt.scatter(temp,icecream_sales,s=100,marker="*",color="r",edgecolor='b',
           linewidths=1.5,alpha=0.5)
plt.title('Temp Vs. Iceream Sales')
plt.xlabel('Temprature')
plt.ylabel('Ice Cream Sales')
plt.show()
```

## 7. Set Different Colors For Different Dots

```python
color=[x for x in range(2,9)]
```

```python
plt.scatter(temp,icecream_sales,s=150,marker="*",c=color,
            edgecolor='b',linewidths=0.5,alpha=None)
plt.title('Temp Vs. Iceream Sales')
plt.xlabel('Temprature')
plt.ylabel('Ice Cream Sales')
plt.show()
```

## 8. Use In-built Color Maps

https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
plt.scatter(temp,icecream_sales,s=150,marker="*",c=color,
            edgecolor='b',linewidths=0.5,alpha=None,cmap='Reds')
plt.title("Temp Vs. Ice Cream Sales")
plt.xlabel('Temprature')
plt.ylabel('Ice Cream Sales')
plt.show()
```

## 9. Add Colorbar

```python
plt.scatter(temp,icecream_sales,s=150,marker="*",c=color,
            edgecolor='b',linewidths=1.5,alpha=None,cmap='Reds')
plt.title('Temp Vs. Iceream Sales')
plt.xlabel('Temprature')
plt.ylabel('Ice Cream Sales')
plt.colorbar()
plt.show()
```

## 10. Customize The Limit of Colorbar

```python
plt.scatter(temp,icecream_sales,s=150,marker="*",c=color,
            edgecolor='b',linewidths=1.5,alpha=None,cmap='Reds',
           vmin=0,vmax=10)
plt.title('Temp Vs. Iceream Sales')
plt.xlabel('Temprature')
plt.ylabel('Ice Cream Sales')
c_bar=plt.colorbar()
c_bar.set_label("Temp Vs. Ice Cream Sales")
plt.show()
```

## 11. Set Different Size For Different Dots

```python
size=np.random.randint(low=500,high=2000,size=7)
```

```python
plt.scatter(temp,icecream_sales,s=size,marker="*",c=color,
            edgecolor='k',linewidths=0.5,alpha=1)
plt.title('Temp Vs. Iceream Sales')
plt.xlabel('Temprature')
plt.ylabel('Ice Cream Sales')
plt.show()
```

## 12. Draw Two Scatter Plots In One Figure

```python
plt.scatter(temp,icecream_sales,s=100,marker="*",c='y',
            edgecolor='k',linewidths=0.5,alpha=1)
plt.scatter(temp,tea_sales,s=100,marker="*",c='r',
            edgecolor='k',linewidths=0.5,alpha=1)
plt.title('Temp Vs. Iceream Sales')
plt.xlabel('Temprature')
plt.ylabel('Ice Cream Sales')
plt.show()
```

## 13. Add Legend To The Plot

```python
plt.scatter(temp,icecream_sales,s=100,marker="*",
            color='y',edgecolor='k',linewidths=0.5,alpha=None,
            label="Icecream Sales")
plt.scatter(temp,tea_sales,s=100,marker="8",
            color='r',edgecolor='k',linewidths=0.5,alpha=None,
            label="Tea Sales")
plt.title('Iceream Sales Vs. Tea Sales')
plt.xlabel('Temptrature')
plt.ylabel('Quantity')
plt.colorbar()
plt.legend()
plt.show()
```

## 14. Change Figure Size & Color

```python
plt.figure(figsize=(10,4),facecolor='lightgrey',edgecolor='black',linewidth=4)
plt.scatter(temp,icecream_sales,s=100,marker="*",
            color='y',edgecolor='k',linewidths=0.5,alpha=None,
            label="Icecream Sales")
plt.scatter(temp,tea_sales,s=100,marker="8",
            color='r',edgecolor='k',linewidths=0.5,alpha=None,
            label="Tea Sales")
plt.title('Iceream Sales Vs. Tea Sales')
plt.xlabel('Temptrature')
plt.ylabel('Quantity')
plt.colorbar()
plt.legend()
plt.show()
```
