```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
grades = ['A','B','C','D']
test1=[3,13,10,4]
test2=[5,14,6,5]
test3=[7,15,5,3]
```

## 1. Draw A Bar Plot

```python
plt.bar(grades,test1)
plt.show()
```

## 2. Give Title, Xlabel, and Ylabel 

```python
plt.bar(grades,test1)
plt.title('Bar Plot')
plt.xlabel('Grades')
plt.ylabel('Number of Students')
plt.show()
```

## 3. Change Font Size of Title, Xlabel, and Ylabel

```python
plt.bar(grades,test1)
plt.title('Bar Plot',fontsize=15)
plt.xlabel('Grades',fontsize=10)
plt.ylabel('Number of Grades',fontsize=10)
plt.show()
```

## 4. Set Width of The Bars & Align Bars To Center

```python
plt.bar(grades,test1,width=0.3,align='center')
plt.title('Bar Plot')
plt.xlabel('Grades')
plt.ylabel('Number of Grades')
plt.show()
```

## 5. Change Color of The Bars & Also Change Edges Color

```python
plt.bar(grades,test1,width=0.4,align='center',color='y',edgecolor='k')
plt.title('Bar Plot')
plt.xlabel('Grades')
plt.ylabel('Number of Grades')
plt.show()
```

## 6. Change  Line Styles & Line Width of The Bars

```python
plt.bar(grades,test1,width=0.4,align='center',color='red',edgecolor='blue',
       linestyle='--',linewidth=3)
plt.title('Bar Plot')
plt.xlabel('Grades')
plt.ylabel('Number of Grades')
plt.show()
```

## 7. Draw Bar Plots For Test1 and Test2

```python
index=np.arange(len(grades))
index
```

```python
plt.bar(index,test1,width=0.4,align='center',color='red',edgecolor='black',
       linestyle='-.',linewidth=1.5)
plt.bar(index+0.3,test2,width=0.4,align='center',color='b',edgecolor='black',
       linestyle='-.',linewidth=1.5)
plt.title('Bar Plot')
plt.xlabel('Grades')
plt.ylabel('Number of Grades')
plt.xticks(index+0.2,grades)
plt.show()
```

## 8. Draw Bar Plots For Test1, Test2 & Test3

```python
plt.figure(figsize=(7,5))
plt.bar(index,test1,width=0.3,align='center',color='red',edgecolor='black',
       linestyle='-.',linewidth=1.5)
plt.bar(index+0.3,test2,width=0.3,align='center',color='b',edgecolor='black',
       linestyle='-.',linewidth=1.5)
plt.bar(index+0.6,test2,width=0.3,align='center',color='y',edgecolor='black',
       linestyle='-.',linewidth=1.5)
plt.title('Bar Plot')
plt.xlabel('Grades')
plt.ylabel('Number of Grades')
plt.xticks(index+0.3,grades)
plt.show()
```

## 9. Add Legend 

```python
plt.figure(figsize=(7,5))
plt.bar(index,test1,width=0.3,align='center',color='red',edgecolor='black',
       linestyle='-.',linewidth=1.5,label="Test1")
plt.bar(index+0.3,test2,width=0.3,align='center',color='b',edgecolor='black',
       linestyle='-.',linewidth=1.5,label="Test2")
plt.bar(index+0.6,test2,width=0.3,align='center',color='y',edgecolor='black',
       linestyle='-.',linewidth=1.5,label="Test3")
plt.title('Bar Plot')
plt.xlabel('Grades')
plt.ylabel('Number of Grades')
plt.xticks(index+0.3,grades,rotation=45)
plt.legend()
plt.show()
```

## 10. Change Figure Size & Color

```python
plt.figure(figsize=(7,5),facecolor='lightgray',edgecolor='black',linewidth=2)
plt.bar(index,test1,width=0.3,align='center',color='red',edgecolor='black',
       linestyle='-.',linewidth=1.5,label="Test1")
plt.bar(index+0.3,test2,width=0.3,align='center',color='b',edgecolor='black',
       linestyle='-.',linewidth=1.5,label="Test2")
plt.bar(index+0.6,test2,width=0.3,align='center',color='y',edgecolor='black',
       linestyle='-.',linewidth=1.5,label="Test3")
plt.title('Bar Plot')
plt.xlabel('Grades')
plt.ylabel('Number of Grades')
plt.xticks(index+0.3,grades,rotation=45)
plt.legend()
plt.show()
```
