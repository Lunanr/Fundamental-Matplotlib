```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 1. Draw A Line Plot

```python
x=[3,4,5,10]
y=[2,6,8,9]
plt.plot(x,y)
plt.show()
```

## 2. Give Title, Xlabel and Ylabel

```python
x=[3,4,5,10]
y=[2,6,8,9]
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

## 3. Change Font Size of Title, Xlabel and Ylabel and also Change location

```python
x=[3,4,5,10]
y=[2,6,8,9]
plt.plot(x,y)
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.show()
```

## 4. Change Color & Width of The Line

```python
x=[3,4,5,10]
y=[2,6,8,9]
plt.plot(x,y,color='b',linewidth=3)
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.show()
```

## 5. Change Marker, Marker Size &  Line Style

```python
x=[3,4,5,10]
y=[2,6,8,9]
plt.plot(x,y,color='b',linewidth=3,marker='D',markersize=10,
        markerfacecolor='red',markeredgecolor='red',
        linestyle='--')
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.show()
```

```python
x=[3,4,5,10]
y=[2,6,8,9]
plt.plot(x,y,'b--<',linewidth=3,markersize=10,
        markerfacecolor='red',markeredgecolor='red',
        )
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.show()
```

## 6. Add Legend To Your Graph Also Change Location

```python
x=np.random.randint(low=1,high=15,size=15)
y=np.random.randint(low=1,high=10,size=15)
plt.plot(x,y,'b--<',linewidth=3,markersize=10,
        markerfacecolor='red',markeredgecolor='red',
        )
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.legend(["A simple line plot"],loc=0)
plt.show()
```

## 8. Xlim And Ylim

```python
x=np.random.randint(low=1,high=15,size=15)
y=np.random.randint(low=1,high=10,size=15)
plt.plot(x,'b--<',linewidth=3,markersize=10,
        markerfacecolor='red',markeredgecolor='red',label="Line 1")
plt.plot(y,'g--<',linewidth=3,markersize=10,
        markerfacecolor='red',markeredgecolor='red',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.xlim(0,20)
plt.ylim(0,15)
plt.legend()
plt.show()
```

## 9. Change X and Y limit Using axis Method

```python
x=np.random.randint(low=1,high=15,size=15)
y=np.random.randint(low=1,high=10,size=15)
plt.plot(x,'b--<',linewidth=3,markersize=10,
        markerfacecolor='red',markeredgecolor='red',label="Line 1")
plt.plot(y,'g--<',linewidth=3,markersize=10,
        markerfacecolor='red',markeredgecolor='red',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
# xmin, xmax, ymin, ymax
plt.axis([0,20,0,15])
plt.legend()
plt.show()
```

## 10. Change X and Y limit Using Xticks And Yticks.

```python
x_ticks=[x for x in range(0,20,2)]
y_ticks=[y for y in range(0,15,3)]
```

```python
x=np.random.randint(low=1,high=15,size=15)
y=np.random.randint(low=1,high=10,size=15)
plt.plot(x,'b--<',linewidth=3,markersize=10,
        markerfacecolor='red',markeredgecolor='red',label="Line 1")
plt.plot(y,'g--<',linewidth=3,markersize=10,
        markerfacecolor='red',markeredgecolor='red',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.show()
```

## 11. Use Grid Method

```python
x=np.random.randint(low=1,high=15,size=15)
y=np.random.randint(low=1,high=10,size=15)
plt.plot(x,'b--<',linewidth=3,markersize=10,
        markerfacecolor='red',markeredgecolor='red',label="Line 1")
plt.plot(y,'g--<',linewidth=3,markersize=10,
        markerfacecolor='red',markeredgecolor='red',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.grid(color='purple',linestyle='--',linewidth=1)
plt.show()
```

## 12. Change Figure Size

```python
x=np.random.randint(low=1,high=15,size=15)
y=np.random.randint(low=1,high=10,size=15)
plt.figure(figsize=(10,4),facecolor='lightgreen',edgecolor='black',
          linewidth=1)
plt.plot(x,'b--<',linewidth=3,markersize=10,
        markerfacecolor='red',markeredgecolor='red',label="Line 1")
plt.plot(y,'g--<',linewidth=3,markersize=10,
        markerfacecolor='red',markeredgecolor='red',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.grid(color='purple',linestyle='--',linewidth=1)
plt.show()
```

## 13. Save Graph

```python
x=np.random.randint(low=1,high=15,size=15)
y=np.random.randint(low=1,high=10,size=15)
plt.figure(figsize=(10,4),facecolor='lightgreen',edgecolor='black',
          linewidth=1)
plt.plot(x,'b--<',linewidth=3,markersize=10,
        markerfacecolor='red',markeredgecolor='red',label="Line 1")
plt.plot(y,'g--<',linewidth=3,markersize=10,
        markerfacecolor='red',markeredgecolor='red',label="Line 2")
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.grid(color='purple',linestyle='--',linewidth=1)
plt.savefig('Lineplot.png')
plt.show()
```
