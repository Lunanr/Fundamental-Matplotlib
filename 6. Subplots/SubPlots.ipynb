```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('Iris.csv')
data.head()
```

# Subplot ()

## 1. Draw Two Plot Side By Side [Histogram of SepalLength & PetalLength]

```python
data.columns
```

```python
plt.suptitle("Iris Dataset")
plt.subplot(1,2,1)
plt.title("SepalLengthCm")
plt.hist(data.SepalLengthCm,color='red')
plt.xlabel("SepalLengthCm")
plt.ylabel("Count")

plt.subplot(1,2,2)
plt.title("PetalLengthCm")
plt.hist(data.PetalLengthCm)
plt.xlabel("SepalLengthCm")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

## 2. Draw Two Plots Top of Each Other

```python
plt.suptitle("Iris Dataset")
plt.subplot(2,1,1)
plt.title("SepalLengthCm")
plt.hist(data.SepalLengthCm,color='red')
plt.xlabel("SepalLengthCm")
plt.ylabel("Count")

plt.subplot(2,1,2)
plt.title("PetalLengthCm")
plt.hist(data.PetalLengthCm,color='green')
plt.xlabel("SepalLengthCm")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

# Subplots ()

## 1. Draw Two Plot Side By Side [Histogram of SepalLength & PetalLength]

```python
data.columns
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2)
fig.suptitle("Iris Dataset")
ax1.hist(data.SepalLengthCm,color='r')
ax1.set_title("SepalLengthCm")
ax1.set_xlabel("SepalLengthCm")
ax1.set_ylabel("Count")


ax2.hist(data.PetalLengthCm,color='g')
ax2.set_title("PetalLengthCm")
ax2.set_xlabel("PetalLengthCm")
ax2.set_ylabel("Count")


plt.tight_layout()
plt.show()
```

## 2. Draw Two Plots Top of Each Other

```python
fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1)
fig.suptitle("Iris Dataset")
ax1.hist(data.SepalLengthCm,color='r')
ax1.set_title("SepalLengthCm")
ax1.set_xlabel("SepalLengthCm")
ax1.set_ylabel("Count")


ax2.hist(data.PetalLengthCm,color='g')
ax2.set_title("PetalLengthCm")
ax2.set_xlabel("PetalLengthCm")
ax2.set_ylabel("Count")


plt.tight_layout()
plt.show()
```

## 3. Draw Four Plots

```python
data.columns
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2)
fig.suptitle("Iris Dataset")
ax1.hist(data.SepalLengthCm,color='r')
ax1.set_title("SepalLengthCm")
ax1.set_xlabel("SepalLengthCm")
ax1.set_ylabel("Count")

ax2.hist(data.PetalLengthCm,color='g')
ax2.set_title("PetalLengthCm")
ax2.set_xlabel("PetalLengthCm")
ax2.set_ylabel("Count")

ax3.hist(data.SepalWidthCm,color='c')
ax3.set_title("SepalWidthCm")
ax3.set_xlabel("SepalWidthCm")
ax3.set_ylabel("Count")

ax4.hist(data.PetalWidthCm,color='y')
ax4.set_title("PetalWidthCm")
ax4.set_xlabel("PetalWidthCm")
ax4.set_ylabel("Count")

plt.tight_layout()
plt.show()
```

## 4. Sharex and sharey

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,sharex=True,sharey=True)
fig.suptitle("Iris Dataset")
ax1.hist(data.SepalLengthCm,color='r')
ax1.set_title("SepalLengthCm")
ax1.set_xlabel("SepalLengthCm")
ax1.set_ylabel("Count")

ax2.hist(data.PetalLengthCm,color='g')
ax2.set_title("PetalLengthCm")
ax2.set_xlabel("PetalLengthCm")
ax2.set_ylabel("Count")

ax3.hist(data.SepalWidthCm,color='c')
ax3.set_title("SepalWidthCm")
ax3.set_xlabel("SepalWidthCm")
ax3.set_ylabel("Count")

ax4.hist(data.PetalWidthCm,color='y')
ax4.set_title("PetalWidthCm")
ax4.set_xlabel("PetalWidthCm")
ax4.set_ylabel("Count")

plt.tight_layout()
plt.show()
```

## 5. Change Figure Size & Color

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,sharex=True,
                                       sharey=True,figsize=(10,6))
fig.suptitle("Iris Dataset")
fig.set_edgecolor("yellow")

ax1.hist(data.SepalLengthCm,color='r')
ax1.set_title("SepalLengthCm")
ax1.set_xlabel("SepalLengthCm")
ax1.set_ylabel("Count")

ax2.hist(data.PetalLengthCm,color='g')
ax2.set_title("PetalLengthCm")
ax2.set_xlabel("PetalLengthCm")
ax2.set_ylabel("Count")

ax3.hist(data.SepalWidthCm,color='c')
ax3.set_title("SepalWidthCm")
ax3.set_xlabel("SepalWidthCm")
ax3.set_ylabel("Count")

ax4.hist(data.PetalWidthCm,color='y')
ax4.set_title("PetalWidthCm")
ax4.set_xlabel("PetalWidthCm")
ax4.set_ylabel("Count")

plt.tight_layout()
plt.show()
```
