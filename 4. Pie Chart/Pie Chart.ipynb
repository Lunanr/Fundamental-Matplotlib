```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
movie = ['Comedy','Action','Romance','Drama']
votes = [5,6,3,1]
```

## 1. Draw A Pie Chart

```python
plt.pie(votes)
plt.show()
```

## 2. Give Title, Label (Also Change Label Distance), And Legend

```python
plt.pie(votes, labels=movie, labeldistance=1.1)
#Labels must be array
plt.title("Favorite Type of Movie")
plt.legend(loc=3)
plt.show()
```

## 3. Pull out Action and Drama Wedges

```python
plt.pie(votes, labels=movie, labeldistance=1.1,
       explode=[0,0.1,0,0.1])
plt.title("Favorite Type of Movie")
plt.legend(loc=3)
plt.show()
```

## 4. Change The Start Angle To 90 Degree

```python
plt.pie(votes, labels=movie, labeldistance=1.1,
       explode=[0,0.1,0,0.1], startangle=90)
plt.title("Favorite Type of Movie")
plt.legend(loc=3)
plt.show()
```

## 5. Set The Color of Each Wedge

```python
plt.pie(votes, labels=movie, labeldistance=1.1,
       explode=[0,0.1,0,0.1], startangle=90,colors=['g','y','m','b'])
plt.title("Favorite Type of Movie")
plt.legend(loc=3)
plt.show()
```

## 6. Display Percentage & Change Distance Inside Each Wedge

```python
plt.pie(votes, labels=movie, labeldistance=1.1,
       explode=[0,0.1,0,0.1], startangle=0,colors=['g','y','m','b'],
       autopct="%0.1f%%",pctdistance=0.7)
plt.title("Favorite Type of Movie")
plt.legend(loc=3)
plt.show()
```

## 7. Change Radius of The Pie Chart

```python
plt.pie(votes, labels=movie, labeldistance=1.1,
       explode=[0,0.1,0,0.1], startangle=0,colors=['g','y','m','b'],
       autopct="%0.1f%%",pctdistance=0.7,radius=1.2)
plt.title("Favorite Type of Movie")
plt.legend(loc=3)
plt.show()
```

## 8. Set Labels Size of The Pie Chart

```python
plt.pie(votes, labels=movie, labeldistance=1.1,
       explode=[0,0.1,0,0.1], startangle=0,colors=['g','y','m','b'],
       autopct="%0.1f%%",pctdistance=0.6,radius=1.2,
       textprops={'fontsize':12})
plt.title("Favorite Type of Movie")
plt.legend(loc=3)
plt.show()
```

## 9. Set Border of The Wedges

```python
plt.pie(votes, labels=movie, labeldistance=1.1,
       explode=[0,0.1,0,0.1], startangle=0,colors=['g','y','m','b'],
       autopct="%0.1f%%",pctdistance=0.6,radius=1.2,
       textprops={'fontsize':12}, wedgeprops={'edgecolor':'blue',
                                             'linewidth':1,
                                             'linestyle':'-.'})
plt.title("Favorite Type of Movie")
plt.legend(loc=3)
plt.show()
```

## 10. Add A shadow to the Pie Chart

```python
plt.pie(votes, labels=movie, labeldistance=1.1,
       explode=[0,0.1,0,0.1], startangle=0,colors=['g','y','m','b'],
       autopct="%0.1f%%",pctdistance=0.6,radius=1.2,
       textprops={'fontsize':12}, wedgeprops={'edgecolor':'blue',
                                             'linewidth':0.5,
                                             'linestyle':'-.'},
       shadow=True)
plt.title("Favorite Type of Movie")
plt.legend(loc=3)
plt.show()
```

## 11. Change figure Size & Color

```python
plt.figure(figsize=(10,5),edgecolor='black',linewidth=4,facecolor='lightblue')
plt.pie(votes, labels=movie, labeldistance=1.1,
       explode=[0,0.1,0,0.1], startangle=0,colors=['g','y','m','b'],
       autopct="%0.1f%%",pctdistance=0.6,radius=1.2,
       textprops={'fontsize':12}, wedgeprops={'edgecolor':'blue',
                                             'linewidth':0.5,
                                             'linestyle':'-.'},
       shadow=True)
plt.title("Favorite Type of Movie")
plt.legend(loc=3)
plt.show()
```
